```python
def weighted_avg_and_std(values, weights):
    average = np.average(values, weights=weights)
    variance = np.average((values-average)**2, weights=weights)
    return (average, math.sqrt(variance))
```

```python
samples_shear = [np.asarray(samps.parameters) for samps in agg.values("samples")]
slope_samples_shear = [samps.parameters_extract[7] for samps in agg.values("samples")]
weights_shear = [np.asarray(samps.weights) for samps in agg.values("samples")]
slopes_bells =[instance.galaxies.lens.mass.slope for instance in instances_bells]
slope_samples_bells = [samps.parameters_extract[7] for samps in agg_bells.values("samples")]
weights_bells = [np.asarray(samps.weights) for samps in agg_bells.values("samples")]
slopes_shear=[instance.galaxies.lens.mass.slope for instance in instances]

slopes_both = slopes_shear+slopes_bells
samples_both = slope_samples_shear+slope_samples_bells
weights_both = weights_shear+weights_bells
std_slopes_shear = [weighted_avg_and_std(values=samps, weights=weights)[1] for samps, weights, in zip(slope_samples_shear, weights_shear)]
std_slopes_shear_bells = [weighted_avg_and_std(values=samps, weights=weights)[1] for samps, weights, in zip(slope_samples_bells, weights_bells)]


std_slopes_both = [weighted_avg_and_std(values=samps, weights=weights)[1] for samps, weights, in zip(samples_both, weights_both)]
(mu_shear, sigma_shear) = norm.fit(slopes_both)
```

```python
def func(sigma, gamma, delta_gamma, av_gamma):
    num = np.square(gamma - av_gamma)-np.square(sigma)-np.square(delta_gamma)
    denom = np.square(np.square(sigma)+np.square(delta_gamma))
    return np.sum(num/denom)
```

```python
from scipy import optimize

av_sigma_shear = optimize.root(func, x0=0.1, args=(slopes_both, std_slopes_both, mu_shear), method='lm', options={"xtol": 0.00001})
```

```python
def velocity_dispersion_from(*, redshift_0, redshift_1, einstein_radius, cosmology):
    const = constants.c.to("kpc / s")
    angular_diameter_distance_to_redshift_0_kpc = cosmology_util.angular_diameter_distance_to_earth_in_kpc_from(
        redshift=redshift_0,
        cosmology=cosmology
    )
    angular_diameter_distance_to_redshift_1_kpc = cosmology_util.angular_diameter_distance_to_earth_in_kpc_from(
        redshift=redshift_1,
        cosmology=cosmology
    )
    angular_diameter_distance_between_redshifts_kpc = cosmology_util.angular_diameter_distance_between_redshifts_in_kpc_from(
        redshift_0=redshift_0, redshift_1=redshift_1, cosmology=cosmology
    )
    kpc_per_arcsec = cosmology_util.kpc_per_arcsec_from(redshift=redshift_0, cosmology=cosmology)
    einstein_radius_kpc = einstein_radius * kpc_per_arcsec
    velocity_dispersion_kpc = \
        const * np.sqrt(
            (einstein_radius_kpc * angular_diameter_distance_to_redshift_1_kpc) /
            (4 * np.pi * angular_diameter_distance_to_redshift_0_kpc * angular_diameter_distance_between_redshifts_kpc)
        )
    return velocity_dispersion_kpc.to("km/s").value
```

```python
from astropy import constants
from astropy import units as u

sigma_SIE = []
sigma_frac = []
sigma_SIE_bells = []
sigma_frac_bells = []

info_gen = agg.values("info")
infos = [info for info in info_gen]
info_gen_bells = agg_bells.values("info")
infos_bells =[info for info in info_gen_bells]

for info, instance, r in zip(infos, instances, R_Ein):
    sigma = velocity_dispersion_from(redshift_0=info["redshift_lens"], 
                                     redshift_1=info["redshift_source"],
                                     einstein_radius=r,
                                     cosmology=cosmo)
    frac = info["sigma"]/sigma
    sigma_SIE.append(sigma)
    sigma_frac.append(frac)
    
for info, instance, r in zip(infos_bells, instances_bells, R_Ein_bells):
    sigma = velocity_dispersion_from(redshift_0=info["redshift_lens"], 
                                     redshift_1=info["redshift_source"],
                                     einstein_radius=r,
                                     cosmology=cosmo)
    if info["velocity_dispersion"] < 800:
        frac = info["velocity_dispersion"]/sigma
        sigma_SIE_bells.append(sigma)
        sigma_frac_bells.append(frac) 
    else:    
        sigma_SIE_bells.append(float("nan"))
        sigma_frac_bells.append(float("nan"))  
```
